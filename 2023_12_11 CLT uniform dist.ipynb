```python
import numpy as np
from scipy.stats import uniform, norm
import matplotlib.pyplot as plt
```

```python
""" 
In probability theory, the central limit theorem (CLT) establishes that, 
in many situations, for independent and identically distributed random variables, 
the sampling distribution of the standardized sample mean tends towards the standard normal distribution 
even if the original variables themselves are not normally distributed.
"""

# sample univariate data with n>30 points
# and investigate how the distribution of the mean and variance converge to a normal distribution

n = 100
n_rep = 5000

means   = np.zeros(n_rep)
vars    = np.zeros(n_rep)

for i in range(n_rep) :

    univ_sample = uniform.rvs(size=n)

    # plot the histogram of the sample
    # plt.hist(univ_sample, bins=20, density=True)
    # plt.show()

    means[i] = np.mean(univ_sample)
    vars[i] = np.var(univ_sample)

# plot the histogram of the sample means and vars
fig, axes = plt.subplots(1,2, figsize=(15,5))
ax1, ax2 = axes

ax1.hist(means, bins=50, density=True)
ax1.set_title("Histogram of the sample means")

ax2.hist(vars, bins=50, density=True)
ax2.set_title("Histogram of the sample vars")

# plot a fitted normal pdf on top of the histogram
x = np.linspace(0.3, 0.7, 100)
ax1.plot(x, norm.pdf(x, loc=0.5, scale=0.03), 'r-', lw=2, label='norm pdf')

x = np.linspace(0.06, 0.105, 100)
ax2.plot(x, norm.pdf(x, loc=0.083, scale=0.008), 'r-', lw=2, label='norm pdf')

plt.tight_layout()
plt.show()
```
